# Data Exploration (fun)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/train.csv")
df.head(10)
```

```python
print(len(df))
targets = len(df["TARGET"].unique())
print(targets)
```

```python
target_counts = df["TARGET"].value_counts().sort_index()

plt.figure(figsize=(20, 10))
sns.barplot(x=target_counts.index, y=target_counts.values, palette='viridis')
plt.xlabel('Butterfly Classes')
plt.ylabel('Number of Images')
plt.xticks(rotation=90)
plt.show()
```

```python
sorted_counts = target_counts.sort_values(ascending=False)
summary = pd.DataFrame({
    'count': sorted_counts,
    'percentage': (sorted_counts / len(df) * 100).round(2)
})
summary.head(10)
```

```python
image_dir = "data/train_images/train_images/"
sample_images = df["ID"].sample(9, random_state=42).values
plt.figure(figsize=(15, 15))
for i, img_id in enumerate(sample_images):
    img_path = f"{image_dir}{img_id}.jpg"
    img = plt.imread(img_path)
    plt.subplot(3, 3, i + 1)
    plt.imshow(img)
    target = df[df["ID"] == img_id]["TARGET"].values[0]
    plt.title(f"Class: {target}")
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
for img_id in sample_images:
    img_path = f"{image_dir}{img_id}.jpg"
    img = plt.imread(img_path)
    print(f"Image {img_id}: shape {img.shape}, dtype {img.dtype}, min {img.min()}, max {img.max()}")
```
